```python
f = open('BAD TEST Serch Terms.txt', 'r',encoding="utf8")

badTerms = f.read()

print(badTerms)

f.close()
```

```python
print(type(badTerms))
```

```python
def num_of_open_paren(word):
    word = str(word)
    counter = Counter(word)
    return counter['('] - counter[')']

openCount = 0
elementCount = 0
termList = []

for word in badTerms.split():
    openCount+=num_of_open_paren(word)
    if (word != "OR"):
        try: termList[elementCount]  = termList[elementCount] + " " + word
        except: termList.append(word)
    elif (word == "OR" and openCount == 0):
        elementCount+=1
    elif (word == "OR" and openCount >= 1):
        try: termList[elementCount]  = termList[elementCount] + " " + word
        except: termList.append(word)

            
            
splitList = badTerms.split()
newOpenCount = 0

for i in splitList:
    newOpenCount+=num_of_open_paren(i)
    print("num open {} , newOpenCount {}".format(num_of_open_paren(i),newOpenCount))
```
