```python
# math.isnan() throws up an error when it is passed a value that is not a float so 
def checkNaN(value):
    try: return math.isnan(value)
    except: return False

for i, trial in combinedData.iterrows():
    combinedData.loc[i, "Text Precedence"] = combinedData.loc[i, "Text Precedence"].replace(combinedData.loc[i, "Control Number"],combinedData.loc[i, "Final Names"])
    if not checkNaN(combinedData.loc[i, "FILE_PATH"]):
        combinedData.loc[i, "FILE_PATH"] = combinedData.loc[i, "FILE_PATH"].replace(combinedData.loc[i, "Control Number"],combinedData.loc[i, "Final Names"])
```
